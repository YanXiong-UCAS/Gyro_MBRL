#### Read txt file

```python
def read_data(filepath):
    
    # count lines
    file = open(filepath, "r")
    count = len(file.readlines())
    data = np.empty([count, 16])
    file.seek(0)
    
    # read each line as a numpy array
    row = 0
    for x in file:
        data[row] = np.array(x.split('\t')).astype(np.float)
        row += 1
    file.close()
    
    # change time from usec into sec and shift the beginning to 0
    data[:,11] = data[:,11] / 1e6
    data[:,11] -= data[0,11]
    # sometime time reaches its higher bound and drops to another value
    for i in range(1,len(data)):
        if data[i,11] < data[i-1,11]:
            data[:,11] = np.linspace(0, (len(data)-1)*0.05, len(data))
            break
    
    return data

def angle_normalize(x, lim = np.pi):
    return ((x + lim) % (2 * lim)) - lim 

def plot_data(data):
    
    f, axs = plt.subplots(2,2,figsize=(14,7))

    plt.subplot(2,2,1)
    plt.title('Red gimbal angle',fontsize=20)
    plt.xlabel('time [s]',fontsize=16)
    plt.ylabel(r'$\theta$ [deg]',fontsize=16)
    plt.grid()
    plt.plot(data[:,11], data[:,14])
    plt.plot(data[:,11], angle_normalize(data[:,2],180))

    plt.subplot(2,2,2)
    plt.title('Blue gimbal angle',fontsize=20)
    plt.xlabel('time [s]',fontsize=16)
    plt.ylabel(r'$\theta$ [deg]',fontsize=16)
    plt.grid()
    plt.plot(data[:,11], data[:,15])
    plt.plot(data[:,11], angle_normalize(data[:,4],180))
    
    plt.subplot(2,2,3)
    plt.title('Red gimbal velocity',fontsize=20)
    plt.xlabel('time [s]',fontsize=16)
    plt.ylabel(r'$\dot(\theta)$ [rpm]',fontsize=16)
    plt.grid()
    plt.plot(data[:,11], data[:,3])

    plt.subplot(2,2,4)
    plt.title('Blue gimbal velocity',fontsize=20)
    plt.xlabel('time [s]',fontsize=16)
    plt.ylabel(r'$\dot(\theta)$ [rpm]',fontsize=16)
    plt.grid()
    plt.plot(data[:,11], data[:,5])
```

```python
folderpath = 'gyro_data/'
filepaths = ['m0_ep001-010','m0_ep011-020','m0_ep021-030','m0_ep031-040','m0_ep041-050']
filepaths = ['m1_ep001-010','m1_ep011-020','m1_ep021-030','m1_ep031-040','m1_ep041-050']
filepaths = ['m2_ep001-010','m2_ep011-020','m2_ep021-030','m2_ep031-040','m2_ep041-050']
filepaths = ['m3_ep001-010','m3_ep011-020','m3_ep021-030','m3_ep031-040','m3_ep041-050']
filepaths = ['m4_ep001-010','m4_ep011-020','m4_ep021-030','m4_ep031-040','m4_ep041-050']
filepath = folderpath+filepaths[4]
```

#### Make sure data is correct

```python
data = read_data(filepath+'.txt')
plot_data(data)
```

```python
# Make sure we have 150 episode in a txt file
l = []
for i in range(1, len(data)):
    if (data[i][14] != 0 and data[i-1][14] == 0):
        l.append(i)
len(l)
```

#### Convert

```python
# measured m:

# state part 
# [2] red gimbal position [deg]
# [3] red gimbal velocity [rpm]
# [4] blue gimbal position [deg]
# [5] blue gimbal velocity [rpm]
# [7] disk velocity [rpm] 
# [14] red gimbal reference [deg]
# [15] blue gimbal reference [deg]

# action part
# [9] red gimbal voltage [V]
# [10] blue gimbal voltage [V]
```

```python
# 150 = 10 epochs x 15 episode, 111 = 110 step + 1 initial state, 9 = 7 states and 2 actions
sa = np.zeros([150,111,9])
for i in range(150):
    idx = l[i]
    for j in range(111):
        m = data[idx+j]
        sa[i][j][0] = angle_normalize(m[2]/180*np.pi)
        sa[i][j][1] = m[3]/60*2*np.pi
        sa[i][j][2] = angle_normalize(m[4]/180*np.pi)
        sa[i][j][3] = m[5]/60*2*np.pi
        sa[i][j][4] = angle_normalize(m[14]/180*np.pi)
        sa[i][j][5] = angle_normalize(m[15]/180*np.pi)
        sa[i][j][6] = m[7]/60*2*np.pi
        sa[i][j][7] = m[9]/10
        sa[i][j][8] = m[10]/10
```

```python
# Plot again, make sure everything is right

f, axs = plt.subplots(3,2,figsize=(14,14))

plt.subplot(3,2,1)
plt.title('Red gimbal angle',fontsize=20)
plt.xlabel('time [s]',fontsize=16)
plt.ylabel(r'$\theta$ [rad]',fontsize=16)
plt.grid()
plt.plot(sa[0][:,4])
plt.plot(sa[0][:,0])

plt.subplot(3,2,2)
plt.title('Blue gimbal angle',fontsize=20)
plt.xlabel('time [s]',fontsize=16)
plt.ylabel(r'$\theta$ [rad]',fontsize=16)
plt.grid()
plt.plot(sa[0][:,5])
plt.plot(sa[0][:,1])

plt.subplot(3,2,3)
plt.title('Red gimbal velocity',fontsize=20)
plt.xlabel('time [s]',fontsize=16)
plt.ylabel(r'$\dot(\theta)$ [rad/s]',fontsize=16)
plt.grid()
plt.plot(sa[0][:,1])

plt.subplot(3,2,4)
plt.title('Blue gimbal velocity',fontsize=20)
plt.xlabel('time [s]',fontsize=16)
plt.ylabel(r'$\dot(\theta)$ [rad/s]',fontsize=16)
plt.grid()
plt.plot(sa[0][:,2])

plt.subplot(3,2,5)
plt.title('Red gimbal voltage',fontsize=20)
plt.xlabel('time [s]',fontsize=16)
plt.ylabel('voltage [V]',fontsize=16)
plt.grid()
plt.plot(sa[0][:,7])

plt.subplot(3,2,6)
plt.title('Blue gimbal voltage',fontsize=20)
plt.xlabel('time [s]',fontsize=16)
plt.ylabel('voltage [V]',fontsize=16)
plt.grid()
plt.plot(sa[0][:,8])
```

#### Save as a pickle file
