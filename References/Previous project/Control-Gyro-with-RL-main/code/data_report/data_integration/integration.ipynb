```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(-10,10,1000)
y = np.sign(x)/(1+abs(x))

plt.figure(figsize=(10,6))
plt.plot(x,y)
plt.xlabel('Integral',fontsize=24)
plt.ylabel('Normalized integral',fontsize=24)
plt.title('Integral normalization in previous work',fontsize=24)
plt.grid()

plt.savefig('integral')
```

```python
x = np.linspace(-10,10,1000)
y = np.sign(x)*abs(x)/(1+abs(x))

plt.figure(figsize=(10,6))
plt.plot(x,y)
plt.xlabel('Integral',fontsize=24)
plt.ylabel('Normalized integral',fontsize=24)
plt.title('Integral normalization in this work',fontsize=24)
plt.grid()

plt.savefig('integral_correct')
```
