# Gyroscope Effect of Delay

```python
import numpy as np
import os
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def read_data(filepath):
    
    # count lines
    file = open(filepath, "r")
    count = len(file.readlines())
    data = np.empty([count, 16])
    file.seek(0)
    
    # read each line as a numpy array
    row = 0
    for x in file:
        data[row] = np.array(x.split('\t')).astype(np.float)
        row += 1
    file.close()
    
    # change time from usec into sec and shift the beginning to 0
    data[:,11] = data[:,11] / 1e6
    data[:,11] -= data[0,11]
    
    return data
        
def plot_data(data):
    
    f, axs = plt.subplots(1,2,figsize=(14,7))

    plt.subplot(1,2,1)
    plt.title('Red gimbal angle',fontsize=20)
    plt.xlabel('time [s]',fontsize=16)
    plt.ylabel(r'$\theta$ [deg]',fontsize=16)
    plt.grid()
    plt.plot(data[:,11], data[:,12])
    plt.plot(data[:,11], data[:,2])

    plt.subplot(1,2,2)
    plt.title('Blue gimbal angle',fontsize=20)
    plt.xlabel('time [s]',fontsize=16)
    plt.ylabel(r'$\theta$ [deg]',fontsize=16)
    plt.grid()
    plt.plot(data[:,11], data[:,13])
    plt.plot(data[:,11], data[:,4])
```

```python
filepaths = ["data_20201106_1321.txt", 
        "data_20201106_1645.txt",
        "data_20201106_1306.txt",
        "data_20201106_1310.txt"]
```

## data_20201106_1321:
### model: best with 3 hidden layers
### NN runtime: 4300 us
### start position: red up and blue front
### other: not optimized for compilation

```python
data0 = read_data(filepaths[0])
plot_data(data0)
```

## data_20201106_1645:
### model: best with 3 hidden layers
### NN runtime: 850 us
### start position: red up and blue front
### other: optimized with '-O3' flag

```python
data1 = read_data(filepaths[1])
plot_data(data1)
```

## data_20201106_1306:
### model: best with 3 hidden layers
### NN runtime: 600 us
### start position: red up and blue front
### other: change double to float

```python
data2 = read_data(filepaths[2])
plot_data(data2)
```

## The effect of delay

```python
for i in range(data0.shape[0]):
    if data0[i,12] == 60:
        data0[:,11] -= data0[i, 11] - 5
        break;

for i in range(data1.shape[0]):
    if data1[i,12] == 60:
        data1[:,11] -= data1[i, 11] - 5
        break;
        
for i in range(data2.shape[0]):
    if data2[i,12] == 60:
        data2[:,11] -= data2[i, 11] - 5
        break;

plt.figure(figsize = (20,10))
plt.title('Red gimbal position',fontsize=28)
plt.xlabel('time [s]',fontsize=24)
plt.xticks(fontsize = 24)
plt.ylabel(r'$\theta$ [deg]',fontsize=24)
plt.yticks(fontsize = 24)
plt.grid()
plt.plot(data0[:,11], data0[:,12])
plt.plot(data0[:,11], data0[:,2])
plt.plot(data1[:,11], data1[:,2])
plt.plot(data2[:,11], data2[:,2])
plt.xlim([2, 8])
plt.legend(['reference', '4300us', '850us', '600us'],fontsize=24)
# plt.savefig('delay.png')
```

```python
for i in range(data0.shape[0]):
    if data0[i,13] == 60:
        data0[:,11] -= data0[i, 11] - 5
        break;

for i in range(data1.shape[0]):
    if data1[i,13] == 60:
        data1[:,11] -= data1[i, 11] - 5
        break;
        
for i in range(data2.shape[0]):
    if data2[i,13] == 60:
        data2[:,11] -= data2[i, 11] - 5
        break;

plt.figure(figsize = (20,10))
plt.title('Blue gimbal position',fontsize=28)
plt.xlabel('time [s]',fontsize=24)
plt.xticks(fontsize = 24)
plt.ylabel(r'$\phi$ [deg]',fontsize=24)
plt.yticks(fontsize = 24)
plt.grid()
plt.plot(data0[:,11], data0[:,13])
plt.plot(data0[:,11], data0[:,4])
plt.plot(data1[:,11], data1[:,4])
plt.plot(data2[:,11], data2[:,4])
plt.xlim([2, 8])
plt.legend(['reference', '4300us', '850us', '600us'],fontsize=24)
# plt.savefig('delay.png')
```
