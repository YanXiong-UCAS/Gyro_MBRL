# Plot reward functions

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

#### power function

```python
x = np.arange(0,1,0.00001)
plt.plot(x, -x**2)
# plt.xlim([0,0.2])
```

```python
x = np.arange(0,1,0.00001)
plt.plot(x, -x**0.05)
plt.plot(x, -x**0.1)
plt.plot(x, -x**0.5)
plt.plot(x, -x**2)
plt.grid()
plt.title('Power reward functions',fontsize=20)
plt.xlabel('x',fontsize=18)
plt.ylabel(r'$-x^p$',fontsize=18)
plt.legend(['p=0.05','p=0.1','p=0.5','p=2'],fontsize=14)
# plt.savefig('power_curve.png')
```

#### Exponential function

```python
x = np.arange(0,1,0.00001)
plt.plot(x, np.exp(-10*abs(x))-1)
plt.plot(x, np.exp(-20*abs(x))-1)
plt.plot(x, np.exp(-40*abs(x))-1)
plt.grid()
plt.title('Exponential reward functions',fontsize=20)
plt.xlabel('x',fontsize=18)
plt.ylabel(r'$-(1-e^{cx})$',fontsize=18)
plt.legend(['c=-10','c=-20','c=-40'],fontsize=14)
# plt.savefig('exp_curve.png')
```

#### Sum of power and exponential functions

```python
x = np.arange(0,1,0.001)
plt.plot(x, -x**0.1 + np.exp(-40*abs(x))-1)
plt.plot(x, -x**0.1 + np.exp(-80*abs(x))-1)
# plt.xlim([0,0.2])
```
